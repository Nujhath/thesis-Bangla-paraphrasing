```python
import pandas as pd
from datasets import load_dataset
from tqdm import tqdm


def process_tapaco_dataset(dataset, out_file):
    tapaco = []
    # The dataset has only train split.
    for data in tqdm(dataset):
        keys = data.keys()
        tapaco.append([data[key] for key in keys])
    tapaco_df = pd.DataFrame(
        data=tapaco,
        columns=[
            "source",
            "target",
        ],
    )
    tapaco_df.to_csv(out_file, sep="\t", index=None)
    return tapaco_df

tapaco_df = process_tapaco_dataset(pp_dataset,"bn_huggingface.csv")
```

```python
bn_df = pd.read_csv("bn_huggingface.csv",sep="\t")
```

```python
bn_df.drop(bn_df.index[50000:419966], inplace=True)
```

```python
from simpletransformers.t5 import T5Model
from sklearn.model_selection import train_test_split
import sklearn
```

```python
bn_df.columns = ["input_text","target_text"]
bn_df["prefix"] = "rephrase"
```

```python
bn_df
```

```python
train_data,test_data = train_test_split(bn_df,test_size=0.2)
```

```python
train_data
```

```python
from simpletransformers.t5 import T5Model, T5Args
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("google/mt5-small")

model = AutoModelForSeq2SeqLM.from_pretrained("google/mt5-small")
```

```python
model_args = T5Args()
model_args.max_seq_length = 196
model_args.tokenizer = tokenizer
model_args.train_batch_size = 8
model_args.eval_batch_size = 8
model_args.num_train_epochs = 3
model_args.evaluate_during_training = False
model_args.use_multiprocessing = False
model_args.fp16 = False
model_args.save_steps = -1
model_args.save_eval_checkpoints = False
model_args.save_model_every_epoch = False
model_args.no_cache = True
model_args.reprocess_input_data = True
model_args.overwrite_output_dir = True
model_args.num_return_sequences = 5


```

```python
model = T5Model("mt5", "google/mt5-small", args=model_args)
```

```python
model.train_model(train_data, eval_data=test_data)
```

```python
from simpletransformers.t5 import T5Model
from pprint import pprint
import os
```

```python
root_dir = os.getcwd()
trained_model_path = os.path.join(root_dir,"outputs")
```

```python
args = {
    "overwrite_output_dir": True,
    "max_seq_length": 256,
    "max_length": 50,
    "top_k": 50,
    "top_p": 0.95,
    "num_return_sequences": 1,
}
```

```python
trained_model = T5Model("mt5",trained_model_path,args=args)
```

```python
prefix = "rephrase"
pred = trained_model.predict([f"{prefix}: আমেরিকান এয়ারলাইনসগুলো জানিয়েছে, সব ধরনের ফ্লাইটে এই ত্রুটির প্রভাব পড়েছে। "])
print(pred)
```
